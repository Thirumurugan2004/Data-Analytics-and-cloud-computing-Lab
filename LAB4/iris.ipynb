```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Iris dataset from CSV file
iris_df = pd.read_csv("/home/auist/Documents/thiru/LAB4/iris.csv")

# Define a function to remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Remove outliers from the dataset using boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df)
plt.title("Box Plot of Iris Dataset Before Removing Outliers")
plt.xticks(rotation=45)
plt.show()

# Remove outliers from each numeric column using the IQR method
numeric_columns = iris_df.select_dtypes(include=['float64', 'int64']).columns
for column in numeric_columns:
    iris_df = remove_outliers_iqr(iris_df, column)

# Plot the boxplot after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df)
plt.title("Box Plot of Iris Dataset After Removing Outliers")
plt.xticks(rotation=45)
plt.show()
```
